# 第一题：：爬取免费代理ip网站中的代理ip，并验证爬下来的ip是否可以使用。

    1、获取免费的代理ip，西刺免费代理ip、89免费代理ip等(可以根据学习情况选择其中一个或者多个网站进行抓取)。

    2、将获取的代理ip，用来访问其他页面，验证是否可用

    3、将可以用的代理ip 保存到本地

```python
from selenium import webdriver
from lxml import etree
import requests
import json
import time 
```

```python
base = 'https://www.xicidaili.com/nt/'
```

```python
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.75 Safari/537.36'
}
ip=[]
port=[]
protocal=[]
for i in range(1,5):
    url = base + str(i)
    r = requests.get(url,headers=headers)
    data = etree.HTML(r.text)
    ip += data.xpath('//tr[@class="odd"]/td[2]/text()')
    port += data.xpath('//tr[@class="odd"]/td[3]/text()')
    protocal += data.xpath('//tr[@class="odd"]/td[6]/text()')
# print('The ip is {} :{}'.format(ip,port))
```

```python
print('{}\n{}\n{}\n'.format(ip,port,protocal))
```

```python
valid = []
for i in range(len(ip)):
#     proxy={"{}".format(protocal[i]):"{}://{}:{}".format(protocal[i],ip[i],port[i])}
    proxy={'https':'{}://{}:{}'.format(protocal[i],ip[i],port[i])}
    print(proxy)    
    try:
        r= requests.get('https://www.baidu.com/',proxies = proxy,timeout=10)
        print('{}://{}:{} is healthy'.format(protocal[i],ip[i],port[i]))
        valid.append('{}://{}:{}'.format(protocal[i],ip[i],port[i]))        
    except:
        print('{}://{}:{} is bad!'.format(protocal[i],ip[i],port[i]))
        
```

```python
valid
```

```python
with open(r'IP.txt','a',encoding='utf-8') as f:
    for i in range(len(valid)):
        f.writelines('{}\n'.format(valid[i]))
```

# 第二题：HTML爬虫，爬取链家网中某个地区租房的信息
要求：

1、获取每套房屋的基本信息

2、将结果保存到csv文件
注意：设置延时

```python
from selenium import webdriver
from lxml import etree
import requests
import json
import time 
```

这个P 标签很大。。 文字是在 p2 里

```python
#只爬取了前5页
url_init = 'https://bj.lianjia.com/'
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.75 Safari/537.36'
}
s= requests.Session()
s.get(url_init,headers=headers) #请求首页获取cookies
cookies = s.cookies #为此次获取的cookies
apart=[]
disct=[]
area=[]
info =[]
rent = []
for i in range(1,6):
    url = 'https://bj.lianjia.com/zufang/pg{}/#contentList'.format(str(i))
    r = requests.get(url=url, headers=headers)
    data = etree.HTML(r.text)
    apart +=(data.xpath('//*[@id="content"]/div[1]/div[1]/div/div/p[1]/a/text()'))
    disct += data.xpath('//*[@id="content"]/div[1]/div[1]/div/div/p[2]/a[1]/text()')
    area += data.xpath('//div[@class="content__list"]/div/div/p[2]/text()[5]')
    info += data.xpath('//div[@class="content__list"]/div/div/p[2]/text()[6]')
    rent += data.xpath('//*[@id="content"]/div[1]/div[1]/div/div/span/em/text()')
    time.sleep(10)    
```

```python
apart
```

```python
def remove(raw):
    result=[]
    for i in range(len(raw)):
        temp = raw[i].replace(" ",'')
        fine = temp.replace("\n","")
        result.append(fine)
    return result
```

```python
apart_result=remove(apart)
disct_result=remove(disct)
area_result=remove(area)
info_result=remove(info)
rent_result=remove(rent)
```

```python
apart_result
```

```python
result = []
# result.append(apart[1])
for j in range(len(apart_result)):
   result.append('{},{},{},{},{}\n'.format(apart_result[j],disct_result[j],area_result[j],info_result[j],rent_result[j]))
```

```python
result
```

```python
with open(r'链家.csv','a',encoding='utf-8') as f:
        f.writelines(result)
```
