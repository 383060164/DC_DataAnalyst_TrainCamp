#### 1st try

```python
for i in range(10):
    page = i*30
    page = 0
    url = 'https://image.so.com/zjl?ch=video&sn='+\
    str(page) + '&listtype=new&temp=1'
    r = requests.get(url)
    data = r.content.decode('unicode_escape')
    
#json 常用于网络数据传输和报春的格式 可以理解为一个字符串化的字典
# inspect --> preview 部分可以看出来是一个json 格式(javascript object notation)
#     json.loads(data)
```

```python
json.loads(data)
#会报错是因为有 1."", 数据里面有没处理好的 \ 类似\t \n 之类 
```

```python
#批量去除 "", 前面是key, 后面value 是空的一些去除
def check(data):
    data_t = data
    data_ss=''
    flag = 0
    length=len(data_t)
    for i, j in enumerate(data_t):
        if i+2>length:
            data_ss+=data_t[i]
            continue
        if flag==0 and data_t[i]+data_t[i+1]=='""' and data_t[i+2]!=','and data_t[i+2]!='}':
            flag = 1
            continue
        if data_t[i]+data_t[i+1]=='""' and flag ==1:
            flag=0
            continue
        data_ss+=data_t[i]
    return data_ss
```
