```python
from tkinter import *
import tkinter as tk
import random
goods={'永生花':20,'磁悬浮球':10,'炫彩水龙头':30,'指尖陀螺':20,'3D发光T恤':15,'星空灯':20,'金丝镂空面具':20}
price_list={'永生花':129,'磁悬浮球':79,'炫彩水龙头':39,'指尖陀螺':69,'3D发光T恤':89,'星空灯':49,'金丝镂空面具':99}
products=['永生花','磁悬浮球','炫彩水龙头','指尖陀螺','3D发光T恤','星空灯','金丝镂空面具']
money = 5000
```

```python
on_hit = False  # 默认初始状态为 False
def hit_me():
    global on_hit
    if on_hit == False:     # 从 False 状态变成 True 状态
        on_hit = True
        var.set('you hit me')   # 设置标签的文字为 'you hit me'
    else:       # 从 True 状态变成 False 状态
        on_hit = False
        var.set('') # 设置 文字为空
```

```python
def list_generate():
    '''商品采购清单'''
    product_list=['永生花','磁悬浮球','炫彩水龙头','指尖陀螺','3D发光T恤','星空灯','金丝镂空面具']
    result={}
    for i in range(5):
        product = random.choice(product_list)
        product_list.remove(product)
        num=random.randint(1,20)
        result.update({product:num})
    print("采购清单：{}\n".format(result))
    return result


def transaction():
    '''接待客人函数'''
    global money, goods
    benefit=0
    buy_list=list_generate()
    for i in buy_list:
        if i not in goods.keys(): #如果没有库存
            print("%s 库存没有，请去其他店铺采购！" %i)
            continue
        if goods[i] - buy_list[i]>=0:
            temp = buy_list[i]*price_list[i]
            benefit+=temp
#             res=goods[i]-buy_list[i]
#             print(res)
            goods.update({i:goods[i]-buy_list[i]})
        else:
            lack = buy_list[i]-goods[i]
            print("{} 库存缺少 {} ".format(i,lack))  
    money+=benefit
    print('总共价钱： {}\n流动资金：{}\n剩余库存：{}\n'.format(benefit,money,goods))
```

```python
def return_goods():
    '''退货方程'''
    global money
    refund = 0
    while True:
        print('现有库存：\n{}'.format(goods))
        print('请选择退货产品： \n 1.永生花 ￥129\n 2.磁悬浮球 ￥79 \n 3.炫彩水龙头 ￥39 \n 4.指尖陀螺 ￥69 \n 5.3D发光T恤 ￥89 \n 6.星空灯 ￥49 \n 7.金丝镂空面具 ￥99\n 0.退出')
        opt = int(input())
        if opt not in range(8):
            print("请选择有效选项：0-7")
            continue
        if opt==0:
            print('退货结束')
            money+=refund
            print('\n总共退货所得：{}, 流动资金还剩余：{}\n'.format(refund,money))
            print('新库存为：{}'.format(goods))
            break
        else:
            print('退货'+str(products[opt-1])+'数量：')
            product=products[opt-1]
            temp = int(input())
            price = price_list[products[opt-1]]
            refund+= (price-2)*temp-10
            goods.update({product:goods[product]-temp})
            print('小计：'+str(refund))
```

```python
def get_goods():
    '''进货方程'''
    global money
    cost = 0
    while True:
        print('现有库存：\n{}'.format(goods))
        print('请选择购买选项： \n 1.永生花 ￥129\n 2.磁悬浮球 ￥79 \n 3.炫彩水龙头 ￥39 \n 4.指尖陀螺 ￥69 \n 5.3D发光T恤 ￥89 \n 6.星空灯 ￥49 \n 7.金丝镂空面具 ￥99\n 0.退出')
        opt = int(input())
        if opt not in range(8):
            print("请选择有效选项：0-7")
            continue
        if opt==0:
            print('购买结束')
            money-=cost
            print('\n总共花费：{}, 流动资金还剩余：{}\n'.format(cost,money))
            print('新库存为：{}'.format(goods))
            break
        else:
            print('购买'+str(products[opt-1])+'数量：')
            product=products[opt-1]
            temp = int(input())
            price = price_list[products[opt-1]]
            cost += (price-0.5)*temp+10
            goods.update({product:goods[product]+temp})
            print('小计：'+str(cost))


            
def exit_game():
    '''退出游戏'''
    print('游戏结束，当前金额:{}'.format(money))
    root.destroy()
```

```python
#Create Tkinter Window Session 类似于tensorflow
root = Tk()
root.geometry("900x700") #window size
root.title("***************Welcome to Grocery store***************")

l = tk.Label(root, 
    text='Welcome to My Grocery Store',    # 标签的文字
    bg='green',     # 背景颜色
    font=('Arial', 12),     # 字体和字体大小
    width=40, height=2  # 标签长宽
    )
l.pack()    # 固定窗口位置

#Button1 
Button1=Button(root, text = '接待客人',fg='Blue',bg='yellow',command = transaction)
Button1.pack(side = LEFT, fill=X)
#Button2 
Button2=Button(root, text = '退货',fg='Blue',bg='Red',command = return_goods)
Button2.pack()
#Button3 
Button3=Button(root, text = '进货',fg='Blue',bg='Purple',command =get_goods)
Button3.pack()
#Button4 
Button4=Button(root, text = '退出游戏',fg='Blue',bg='yellow',command = exit_game)
Button4.pack()

root.mainloop() #Keep the window open 

```
